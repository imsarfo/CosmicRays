```python
#Average of the whole set
#len(data) only tru for the number of items
total=0
for i in range(len(histdata[0])):
  total=total +histdata[0][i]*histdata[1][i+1]/len(data)
print(total)
```

```python
def GetAvg(histdata,begin,end):
    bins = histdata[0]
    ranges= histdata[1]
    bincount=0
    freqcount=0
    for i in range(begin,end):
      bincount+=bins[i]
      freqcount+= bins[i]*(ranges[i]+ranges[i+1])/2
      
    return freqcount/bincount


    
```

```python
GetAvg(histdata,1,5)
```
